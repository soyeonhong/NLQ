```python
from pathlib import Path
import json

p_cap_dir = Path('/data/soyeonhong/GroundVQA/llava-v1.6-34b/global')
```

```python

valid_video_id_llava = [video_id.stem for video_id in Path('/data/soyeonhong/GroundVQA/llava-v1.6-34b/global').glob('*.pt')]
annotations = json.loads(Path('/data/soyeonhong/GroundVQA/data/unified/annotations.NLQ_val.json').read_text())
valid_annotations = [annotation for annotation in annotations if annotation['video_id'] in valid_video_id_llava]
```

```python
for annotation in valid_annotations:
    video_id = annotation['video_id']
    start_time = annotation['clip_start_sec']
    end_time = annotation['clip_end_sec']
    
    cap_dir = p_cap_dir / f'{video_id}.json'
    
    cap_data = json.loads(cap_dir.read_text())['answers']
    
    print(len(cap_data), start_time, end_time)
```
